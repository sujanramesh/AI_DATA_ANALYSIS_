```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_numeric_distribution(data: pd.Series, column_name: str):
    """
    Analyzes the distribution of a numeric Pandas Series using histograms and descriptive statistics.

    Args:
        data (pd.Series): The numeric column to analyze.
        column_name (str): The name of the column for labeling purposes.
    """
    print(f"Analyzing distribution of column: '{column_name}'")
    print("-" * 30)

    # 1. Histogram
    plt.figure(figsize=(10, 6))
    sns.histplot(data, kde=True)  # kde=True adds a kernel density estimate line
    plt.title(f'Distribution of {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.show()

    # 2. Descriptive Statistics
    print("\nDescriptive Statistics:")
    print(data.describe())

    # 3. Additional Shape Statistics
    print("\nSkewness and Kurtosis:")
    print(f"Skewness: {data.skew():.2f}")
    print(f"Kurtosis: {data.kurt():.2f}")

    # Interpretation of Skewness and Kurtosis (basic)
    if data.skew() > 0:
        print(f"\n{column_name} is positively skewed (tail extends to the right).")
    elif data.skew() < 0:
        print(f"\n{column_name} is negatively skewed (tail extends to the left).")
    else:
        print(f"\n{column_name} is approximately symmetric.")

    if data.kurt() > 0:
        print(f"{column_name} has higher kurtosis (more peaked, heavier tails) than a normal distribution.")
    elif data.kurt() < 0:
        print(f"{column_name} has lower kurtosis (flatter peak, thinner tails) than a normal distribution.")
    else:
        print(f"{column_name} has a kurtosis similar to a normal distribution.")

if __name__ == '__main__':
    # Example Usage:
    # Create a sample Pandas Series (replace with your actual data)
    data = pd.Series([2, 3, 3, 4, 4, 4, 5, 5, 6, 10, 12])
    column_name = "Sample Data"

    analyze_numeric_distribution(data, column_name)

    # Example with a DataFrame:
    data = {'Values': [15, 20, 20, 25, 30, 35, 40, 40, 45, 60]}
    df = pd.DataFrame(data)
    analyze_numeric_distribution(df['Values'], 'Values')
```
