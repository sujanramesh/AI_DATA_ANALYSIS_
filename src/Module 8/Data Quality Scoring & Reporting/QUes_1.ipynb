### Visualize Basic DQI with Bar Plot

**Description**: Create a bar plot for DQI and errors in a dataset.

```python
# Write your code from here
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read data from CSV
file_path = 'swiggy.csv'  # Replace with your CSV path
df = pd.read_csv(file_path)

# Step 2: Identify errors as missing values (NaNs)
missing_counts = df.isna().sum()
total_rows = len(df)

# Step 3: Calculate DQI per column and overall
dqi_per_column = 1 - (missing_counts / total_rows)
overall_dqi = dqi_per_column.mean()

print("DQI per column:")
print(dqi_per_column)
print(f"\nOverall DQI: {overall_dqi:.4f}")

# Step 4: Visualize Basic DQI with Bar Plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar plot for DQI
dqi_per_column.plot(kind='bar', color='skyblue', ax=ax1)
ax1.set_ylabel('DQI (1 - Missing Ratio)')
ax1.set_ylim(0, 1.05)
ax1.set_title('Data Quality Index (DQI) and Number of Errors per Column')

# Add second y-axis for missing counts
ax2 = ax1.twinx()
missing_counts.plot(kind='line', color='red', marker='o', linewidth=2, ax=ax2)
ax2.set_ylabel('Number of Missing Values (Errors)')
ax2.set_ylim(0, missing_counts.max() + 5)

plt.show()
```
