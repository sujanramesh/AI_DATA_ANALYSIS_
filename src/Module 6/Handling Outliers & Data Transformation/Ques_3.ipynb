```python
# Question: Power Transformation (Box-Cox) Method
# Description: Use Box-Cox transformation to stabilize variance and make the data more normal.


```

```python
import pandas as pd
from sklearn.preprocessing import PowerTransformer
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset with skewed data (positive values only)
data = {
    'Income': [1000, 1500, 2000, 2500, 3000, 10000, 15000, 18000, 25000, 30000]
}

df = pd.DataFrame(data)

# Visualize original distribution
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.histplot(df['Income'], kde=True, color='coral')
plt.title("Original Income Distribution")

# Apply Box-Cox Transformation using PowerTransformer
pt = PowerTransformer(method='box-cox', standardize=False)
df['Income_BoxCox'] = pt.fit_transform(df[['Income']])

# Visualize transformed distribution
plt.subplot(1, 2, 2)
sns.histplot(df['Income_BoxCox'], kde=True, color='seagreen')
plt.title("Box-Cox Transformed Distribution")

plt.tight_layout()
plt.show()

# Optional: Show skewness
print(f"Skewness Before: {df['Income'].skew():.2f}")
print(f"Skewness After: {df['Income_BoxCox'].skew():.2f}")
```
