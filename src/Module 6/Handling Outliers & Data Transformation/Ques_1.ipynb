```python
# Question: Visualizing Outliers Using Boxplots
# Description: Use a boxplot to visually detect outliers in a dataset.


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset with potential outliers
data = {
    'Salary': [40000, 42000, 41000, 45000, 47000, 49000, 120000, 50000, 43000, 44000, 110000]
}

df = pd.DataFrame(data)

# Plot a boxplot to detect outliers
plt.figure(figsize=(8, 5))
sns.boxplot(y=df['Salary'], color='skyblue')
plt.title('Boxplot of Salary')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```
