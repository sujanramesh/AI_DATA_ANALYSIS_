```python
# Question: Winsorization of Data
# Description: Apply Winsorization to a given dataset to handle outliers.

```

```python
import pandas as pd
from scipy.stats.mstats import winsorize
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset with outliers
data = {
    'Salary': [40000, 42000, 41000, 45000, 47000, 49000, 120000, 50000, 43000, 44000, 110000]
}

df = pd.DataFrame(data)

# Plot original data
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.boxplot(y=df['Salary'], color='lightcoral')
plt.title("Original Data (with Outliers)")

# Apply Winsorization: limit 5% from both tails
# `limits=(0.05, 0.05)` means cap the lowest and highest 5%
df['Salary_Winsorized'] = winsorize(df['Salary'], limits=[0.05, 0.05])

# Plot Winsorized data
plt.subplot(1, 2, 2)
sns.boxplot(y=df['Salary_Winsorized'], color='lightgreen')
plt.title("Winsorized Data (Outliers Handled)")

plt.tight_layout()
plt.show()

# Optional: compare statistics
print("Before Winsorization:\n", df['Salary'].describe())
print("\nAfter Winsorization:\n", df['Salary_Winsorized'].describe())
```
