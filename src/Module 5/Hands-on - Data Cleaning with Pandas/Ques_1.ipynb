```python
import pandas as pd
import numpy as np

# -----------------------------
# Task 1: Detect Missing Data
# -----------------------------

# 1. Load the Data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Age': [25, np.nan, 30, 22, np.nan, 29],
    'Gender': ['F', 'M', np.nan, 'M', 'F', np.nan],
    'Income': [50000, 54000, 58000, np.nan, 62000, 61000]
}

df = pd.DataFrame(data)
print("Original Data:\n", df)

# 2. Detect Missing Data
print("\nMissing Values Per Column:\n", df.isnull().sum())
print("\nTotal Missing Values:\n", df.isnull().sum().sum())
print("\nRows with Any Missing Data:\n", df[df.isnull().any(axis=1)])

# -----------------------------
# Task 2: Handle Missing Data by Dropping
# -----------------------------

# 1. Drop Rows with Missing Values
df_dropped = df.dropna()
print("\nAfter Dropping Rows with Missing Values:\n", df_dropped)

# -----------------------------
# Task 3: Handle Missing Data by Imputation
# -----------------------------

# 1. Fill Missing Values
df_imputed = df.copy()

# Fill Age with Mean
df_imputed['Age'].fillna(df_imputed['Age'].mean(), inplace=True)

# Fill Gender with Mode
df_imputed['Gender'].fillna(df_imputed['Gender'].mode()[0], inplace=True)

# Fill Income with Median
df_imputed['Income'].fillna(df_imputed['Income'].median(), inplace=True)

print("\nAfter Imputation:\n", df_imputed)
```
