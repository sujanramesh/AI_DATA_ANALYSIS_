```python
# Title: Basic Data Visualization
# Description: Plot the distribution of one feature using Seaborn.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load a sample dataset from Seaborn (you can replace this with your own DataFrame)
iris = sns.load_dataset('iris')

# Choose the feature you want to visualize the distribution of
feature_to_plot = 'petal_length'

# Create a distribution plot (histogram with KDE) using Seaborn
sns.displot(iris[feature_to_plot], kde=True)

# Add title and labels for clarity
plt.title(f'Distribution of {feature_to_plot}')
plt.xlabel(feature_to_plot)
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
