```python
# Basic Seaborn Bar Plot
# Description: Using Seaborn, plot the same bar plot from Question 1 to compare the sales data.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Product names
products = np.array(['A', 'B', 'C', 'D', 'E'])

# Sales figures (in thousands)
sales = np.array([23, 75, 57, 90, 65])

# Create the bar plot using Seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x=products, y=sales, palette='viridis') # 'viridis' is a color palette

# Add labels and title
plt.xlabel('Product')
plt.ylabel('Sales (in Thousands)')
plt.title('Product Sales (Seaborn)')

# Display the plot
plt.show()
```
