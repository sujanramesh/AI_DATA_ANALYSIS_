```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Heights (in centimeters)
heights = np.array([160, 170, 165, 180, 175])

# Weights (in kilograms)
weights = np.array([55, 65, 60, 80, 76])

# Size of the points
sizes = np.array([10, 15, 13, 18, 12]) * 20 # Multiply by a factor to make the size difference more visible

# Create a Pandas DataFrame (Seaborn often works well with DataFrames)
data = pd.DataFrame({'Height': heights, 'Weight': weights, 'Size': sizes})

# Create the enhanced scatter plot using Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Height', y='Weight', size='Size', data=data, color='coral', marker='o', alpha=0.7)

# Add labels and title
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Relationship between Height and Weight (Seaborn with Size Variation)')

# Add a legend for the size
plt.legend(title='Size', scatterpoints=1, labelspacing=1, handletextpad=0.5)

# Add grid for better readability (optional)
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
