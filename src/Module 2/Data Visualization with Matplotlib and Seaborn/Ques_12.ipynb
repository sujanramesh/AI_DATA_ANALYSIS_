```python
# Advanced Seaborn Plot - Pairplot
# Objective: Use Seaborn to create a pairplot to visualize pairwise relationships in a dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample Dataset (replace with your actual data)
np.random.seed(0)
data = pd.DataFrame({
    'Variable_A': np.random.rand(50) * 10,
    'Variable_B': 2 * np.random.rand(50) * 10 + 5,
    'Variable_C': np.random.randn(50) * 5,
    'Category': np.random.choice(['X', 'Y'], size=50)
})

# Create the pairplot using Seaborn
plt.figure(figsize=(10, 10)) # Adjust figure size as needed
sns.pairplot(data, hue='Category', diag_kind='hist', markers=['o', 's'])

# Add a title to the entire figure
plt.suptitle('Pairwise Relationships in the Dataset (Seaborn)', y=1.02)

# Display the plot
plt.show()
```
