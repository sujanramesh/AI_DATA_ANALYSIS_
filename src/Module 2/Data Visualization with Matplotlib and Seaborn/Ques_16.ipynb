```python
# Combining Plots with Matplotlib
# Objective: Combine multiple plots into a single figure using Matplotlib's subplots.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the plots
x = np.linspace(0, 10, 50)
y_sin = np.sin(x)
y_cos = np.cos(x)
random_data = np.random.rand(50) * 10
categories = ['A', 'B', 'C', 'D']
values = [30, 45, 15, 60]

# Create a figure and a set of subplots
# plt.subplots(nrows, ncols, figsize=(width, height))
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Combined Plots using Matplotlib Subplots', fontsize=16, fontweight='bold')

# Plot 1: Sine wave
axes[0, 0].plot(x, y_sin, color='blue', linestyle='-', label='Sine')
axes[0, 0].set_title('Sine Wave')
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('sin(x)')
axes[0, 0].legend()
axes[0, 0].grid(True, linestyle='--', alpha=0.6)

# Plot 2: Cosine wave
axes[0, 1].plot(x, y_cos, color='red', linestyle='--', label='Cosine')
axes[0, 1].set_title('Cosine Wave')
axes[0, 1].set_xlabel('X')
axes[0, 1].set_ylabel('cos(x)')
axes[0, 1].legend()
axes[0, 1].grid(True, linestyle=':', alpha=0.6)

# Plot 3: Scatter plot
axes[1, 0].scatter(x, random_data, color='green', marker='o', alpha=0.7, label='Random Data')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('X')
axes[1, 0].set_ylabel('Random Value')
axes[1, 0].legend()
axes[1, 0].grid(True, linestyle='-.', alpha=0.6)

# Plot 4: Bar plot
axes[1, 1].bar(categories, values, color='purple', alpha=0.7, label='Bar Data')
axes[1, 1].set_title('Bar Plot')
axes[1, 1].set_xlabel('Category')
axes[1, 1].set_ylabel('Value')
axes[1, 1].legend()
axes[1, 1].grid(axis='y', linestyle='-', alpha=0.6)

# Adjust layout to prevent overlapping subplots
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust rect to make space for suptitle

# Display the combined plot
plt.show()
```
