```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)

# Customization parameters
line1_color = 'teal'
line1_style = '--'
line1_width = 2
line1_marker = 'o'
line1_markersize = 8
line1_label = 'Sine Wave'

line2_color = 'coral'
line2_style = '-.'
line2_width = 1.5
line2_marker = '*'
line2_markersize = 10
line2_label = 'Cosine Wave'

title_font = {'family': 'monospace', 'color': 'navy', 'size': 18, 'weight': 'bold'}
axis_label_font = {'family': 'sans-serif', 'color': 'darkred', 'size': 14, 'style': 'italic'}
legend_font_size = 12

# Create the line plot with customized styles
plt.figure(figsize=(12, 7))

plt.plot(x, y1, color=line1_color, linestyle=line1_style, linewidth=line1_width,
         marker=line1_marker, markersize=line1_markersize, label=line1_label)

plt.plot(x, y2, color=line2_color, linestyle=line2_style, linewidth=line2_width,
         marker=line2_marker, markersize=line2_markersize, label=line2_label)

# Add labels and title with custom fonts
plt.xlabel('X-axis Value', fontdict=axis_label_font)
plt.ylabel('Y-axis Value', fontdict=axis_label_font)
plt.title('Customized Line Plot', fontdict=title_font)

# Add a legend with custom font size
plt.legend(fontsize=legend_font_size)

# Add grid for better readability
plt.grid(True, linestyle=':', alpha=0.7)

# Add annotations (optional)
plt.annotate('Local Maximum (Sine)', xy=(np.pi/2, 1), xytext=(2, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('Local Minimum (Cosine)', xy=(np.pi, -1), xytext=(4, -1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Customize the appearance of ticks
plt.xticks(fontsize=10, color='dimgray')
plt.yticks(fontsize=10, color='dimgray')

# Add a horizontal line at y=0
plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8)

# Improve layout
plt.tight_layout()

# Display the plot
plt.show()
```
