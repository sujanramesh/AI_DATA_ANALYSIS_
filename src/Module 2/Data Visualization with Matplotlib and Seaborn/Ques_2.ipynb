```python
import matplotlib.pyplot as plt
import numpy as np

# Heights (in centimeters)
heights = np.array([160, 170, 165, 180, 175])

# Weights (in kilograms)
weights = np.array([55, 65, 60, 80, 76])

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(heights, weights, color='coral', marker='o', s=80) # 'o' for circle marker, s for size

# Add labels and title
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Relationship between Height and Weight')

# Add grid for better readability (optional)
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
