```python
# Seaborn Histogram with KDE
# Description: Create a histogram using Seaborn that includes a Kernel Density Estimate (KDE) for marks
# data: [23, 45, 56, 78, 89, 90, 67, 88, 76, 59].
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Marks obtained by students
marks = np.array([23, 45, 56, 78, 89, 90, 67, 88, 76, 59])

# Create the histogram with KDE using Seaborn
plt.figure(figsize=(8, 6))
sns.histplot(marks, bins=5, kde=True, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Marks')
plt.ylabel('Frequency / Density')
plt.title('Distribution of Student Marks with KDE (Seaborn)')

# Add grid for better readability (optional)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
