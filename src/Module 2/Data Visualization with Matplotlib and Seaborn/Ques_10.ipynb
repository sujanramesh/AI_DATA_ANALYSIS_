```python
# Advanced Seaborn Plot with DataFrame
# Description: Create a Seaborn plot using a DataFrame that visualizes two features in a single plot. Use tips
# dataset available in Seaborn's sample datasets.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset from Seaborn
tips = sns.load_dataset('tips')

# Display the first few rows of the DataFrame to understand its structure
print(tips.head())

# Create a scatter plot visualizing 'total_bill' vs 'tip', colored by 'sex'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', size='size', data=tips, alpha=0.7)

# Add labels and title
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.title('Total Bill vs. Tip (Colored by Sex, Style by Smoker, Size by Party Size)')

# Add a legend
plt.legend(title='Sex', loc='upper left') # Adjust legend location if needed

# Display the plot
plt.show()
```
