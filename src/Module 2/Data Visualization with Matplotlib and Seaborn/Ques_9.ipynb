```python
# Matplotlib Subplots for Multiple Data Sets
# Description: Create subplots using Matplotlib to display the bar plot for product sales and the scatter plot for
# height vs. weight side by side.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the bar plot (Product Sales)
products = np.array(['A', 'B', 'C', 'D', 'E'])
sales = np.array([23, 75, 57, 90, 65])

# Data for the scatter plot (Height vs. Weight)
heights = np.array([160, 170, 165, 180, 175])
weights = np.array([55, 65, 60, 80, 76])

# Create a figure and a set of subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot the bar plot on the first subplot (axes[0])
axes[0].bar(products, sales, color='skyblue')
axes[0].set_xlabel('Product')
axes[0].set_ylabel('Sales (in Thousands)')
axes[0].set_title('Product Sales')
axes[0].grid(axis='y', linestyle='--')

# Plot the scatter plot on the second subplot (axes[1])
axes[1].scatter(heights, weights, color='coral', marker='o', s=80)
axes[1].set_xlabel('Height (cm)')
axes[1].set_ylabel('Weight (kg)')
axes[1].set_title('Relationship between Height and Weight')
axes[1].grid(True, linestyle='--', alpha=0.7)

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout()

# Display the plot
plt.show()
```
