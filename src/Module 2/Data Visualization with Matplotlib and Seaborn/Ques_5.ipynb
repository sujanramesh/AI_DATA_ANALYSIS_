```python
# Basic Seaborn Scatter Plot
# Description: Using Seaborn, recreate the scatter plot from Question 2 for height vs. weight.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Heights (in centimeters)
heights = np.array([160, 170, 165, 180, 175])

# Weights (in kilograms)
weights = np.array([55, 65, 60, 80, 76])

# Create the scatter plot using Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x=heights, y=weights, color='coral', marker='o', s=80)

# Add labels and title
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Relationship between Height and Weight (Seaborn)')

# Display the plot
plt.show()
```
