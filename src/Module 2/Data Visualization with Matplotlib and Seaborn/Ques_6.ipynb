```python
import matplotlib.pyplot as plt
import numpy as np

# Product names
products = np.array(['A', 'B', 'C', 'D', 'E'])

# Sales figures (in thousands)
sales = np.array([23, 75, 57, 90, 65])

# Customization parameters
bar_color = 'purple'
grid_color = 'gray'
title_font = {'family': 'serif', 'color': 'darkblue', 'size': 16, 'weight': 'bold'}
axis_label_font = {'family': 'sans-serif', 'color': 'darkgreen', 'size': 12}

# Create the bar plot
plt.figure(figsize=(10, 7)) # Adjust figure size

plt.bar(products, sales, color=bar_color)

# Add grid
plt.grid(axis='y', linestyle='-', color=grid_color, alpha=0.7)

# Add labels and title with custom fonts
plt.xlabel('Product', fontdict=axis_label_font)
plt.ylabel('Sales (in Thousands)', fontdict=axis_label_font)
plt.title('Customized Product Sales', fontdict=title_font)

# Improve layout to prevent labels from overlapping
plt.tight_layout()

# Display the plot
plt.show()
```
