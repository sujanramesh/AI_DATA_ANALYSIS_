```python
import pandas as pd
import numpy as np

# --- Part 1: Handle Missing Values & Duplicates ---

print("\n--- Part 1: Handle Missing Values & Duplicates ---")

# 1. Load the Data:
print("\n1. Load the Data:")
data_mv_dup = {'ID': [1, 2, 3, 4, 5, 2, 6, np.nan],
               'Name': ['Alice', 'Bob', np.nan, 'Charlie', 'David', 'Bob', 'Eve', 'Alice'],
               'Age': [25, 30, 28, np.nan, 35, 30, 22, 25],
               'City': ['New York', 'London', 'Paris', 'London', 'Tokyo', 'London', 'Paris', np.nan]}
df_mv_dup = pd.DataFrame(data_mv_dup)
print("Original DataFrame with Missing Values and Duplicates:")
print(df_mv_dup)

# 2. Handling Missing Values
print("\n2. Handling Missing Values:")
#     1. Identify Missing Values:
print("\n   1. Identify Missing Values:")
print("Missing values count per column:")
print(df_mv_dup.isnull().sum())
print("\nDataFrame with boolean mask for missing values:")
print(df_mv_dup.isnull())

#     2. Fill Missing Values:
print("\n   2. Fill Missing Values:")
# Filling 'Name' with 'Unknown'
df_mv_dup['Name'].fillna('Unknown', inplace=True)
# Filling 'Age' with the mean
df_mv_dup['Age'].fillna(df_mv_dup['Age'].mean(), inplace=True)
# Filling 'City' with the mode
df_mv_dup['City'].fillna(df_mv_dup['City'].mode()[0], inplace=True)
# Filling 'ID' with a placeholder value (be cautious with ID filling)
df_mv_dup['ID'].fillna(-1, inplace=True)

print("\nDataFrame after filling missing values:")
print(df_mv_dup)

# 3. Handling Duplicates
print("\n3. Handling Duplicates:")
#     1. Identify Duplicates:
print("\n   1. Identify Duplicates:")
duplicates = df_mv_dup.duplicated()
print("Boolean Series indicating duplicate rows:")
print(duplicates)
print("\nNumber of duplicate rows:", duplicates.sum())
print("\nDuplicate rows:")
print(df_mv_dup[duplicates])

#     2. Remove Duplicates:
print("\n   2. Remove Duplicates:")
df_no_duplicates = df_mv_dup.drop_duplicates()
print("\nDataFrame after removing duplicate rows:")
print(df_no_duplicates)

# 4. Combined Practice on a New Dataset
print("\n4. Combined Practice on a New Dataset:")
#     1. New Sample Data:
print("\n   1. New Sample Data:")
new_data = {'Item': ['Apple', np.nan, 'Banana', 'Orange', 'Apple', 'Grape', 'Banana', np.nan],
            'Quantity': [10, 5, np.nan, 8, 10, 12, 5, 7],
            'Price': [1.0, 2.5, 0.8, 1.2, 1.0, 1.5, 0.8, np.nan]}
new_df = pd.DataFrame(new_data)
print("New DataFrame with Missing Values and Potential Duplicates:")
print(new_df)

#     2. Handling Missing Values:
print("\n   2. Handling Missing Values:")
new_df['Item'].fillna('Unknown', inplace=True)
new_df['Quantity'].fillna(new_df['Quantity'].median(), inplace=True)
new_df['Price'].fillna(new_df['Price'].mean(), inplace=True)
print("\nNew DataFrame after filling missing values:")
print(new_df)

#     3. Remove Duplicates:
print("\n   3. Remove Duplicates:")
new_df_no_duplicates = new_df.drop_duplicates()
print("\nNew DataFrame after removing duplicate rows:")
print(new_df_no_duplicates)

# --- Part 2: Apply Standardization & Formatting Rules ---

print("\n\n--- Part 2: Apply Standardization & Formatting Rules ---")

# 1. Standardize Text Data
print("\n1. Standardize Text Data:")
#     1. Convert All Names to Lowercase:
print("\n   1. Convert All Names to Lowercase:")
data_names = {'Name': ['Alice Smith', 'BOB JOHNSON', 'charlie Brown', 'David Lee']}
df_names = pd.DataFrame(data_names)
print("Original Names DataFrame:")
print(df_names)
df_names['Name_Lower'] = df_names['Name'].str.lower()
print("\nNames DataFrame with lowercase names:")
print(df_names)

# 2. Format Numerical Data
print("\n2. Format Numerical Data:")
#     1. Round Age Column to the Nearest Integer:
print("\n   1. Round Age Column to the Nearest Integer:")
data_ages = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25.3, 30.8, 28.1]}
df_ages = pd.DataFrame(data_ages)
print("Original Ages DataFrame:")
print(df_ages)
df_ages['Age_Rounded'] = df_ages['Age'].round().astype(int)
print("\nAges DataFrame with rounded ages (as integers):")
print(df_ages)

# 3. Combined Practice on Another Dataset
print("\n3. Combined Practice on Another Dataset:")
#     1. New Sample Data:
print("\n   1. New Sample Data:")
product_data = {'Product': ['Laptop Pro', 'Smart Watch', 'LAPTOP pro', 'Wireless Mouse', 'smart watch'],
                'Price': [1200.50, 250.75, 1200.50, 30.00, 250.75]}
df_products = pd.DataFrame(product_data)
print("Original Products DataFrame:")
print(df_products)

#     2. Standardize Product Names:
print("\n   2. Standardize Product Names:")
df_products['Product_Standardized'] = df_products['Product'].str.lower().str.strip()
print("\nProducts DataFrame with standardized product names:")
print(df_products)

#     3. Format Prices to Two Decimal Places:
print("\n   3. Format Prices to Two Decimal Places:")
df_products['Price_Formatted'] = df_products['Price'].map('{:.2f}'.format)
print("\nProducts DataFrame with prices formatted to two decimal places (as strings):")
print(df_products)
```
